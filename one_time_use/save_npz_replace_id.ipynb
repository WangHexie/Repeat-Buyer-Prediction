```python
import numpy as np
from basic_function import *
```

```python
logs_df = get_train_log(None)
```

```python
logs_df = logs_df[logs_df["action_type"] == 2]
```

```python
logs_df.fillna(0, inplace=True)
```

```python
logs_df = logs_df.loc[:, ["user_id","seller_id", "brand_id"]]
```

```python
logs_df = logs_df.astype(int)
```

```python
logs_df.loc[:, ["user_id","seller_id", "brand_id"]]
```

```python
replace = {"user_id": dict(
    zip(set(logs_df.loc[:, "user_id"].values), list(range(len(set(logs_df.loc[:, "user_id"].values)))))),
           "seller_id": dict(
               zip(set(logs_df.loc[:, "seller_id"].values), list(range(len(set(logs_df.loc[:, "seller_id"].values)))))),
           "brand_id": dict(
               zip(set(logs_df.loc[:, "brand_id"].values), list(range(len(set(logs_df.loc[:, "brand_id"].values))))))}
```

```python
replace
```

```python
logs_df.user_id = logs_df.user_id.map(replace["user_id"])
```

```python
logs_df.brand_id = logs_df.brand_id.map(replace["brand_id"])
```

```python
logs_df.seller_id = logs_df.seller_id.map(replace["seller_id"])
```

```python
logs_df
```

```python
train_data = logs_df.loc[:, ["user_id","seller_id", "brand_id"]].values
```
